```python
path = r"C:/Users/iskha/research/process_img/Images/"
img_paths = {'.jpg','.jpeg', 'png'}

imgs = {}

diff = {}

num_imgs = 0

for i in os.listdir(path):
    name, extension = os.path.splitext(i)
    if not extension.lower() in img_paths:
        continue
    
    img = cv2.imread(path+i,1)
    imgs[name] = ProcessImg(img, makeHist = False)
    
    imgs[name].apply_iagcwd()
    imgs[name].apply_contrast_stretch_m1()
    imgs[name].apply_contrast_stretch_m2()
    imgs[name].apply_equalization()
    imgs[name].apply_iagcwd('stretch_m1')
    imgs[name].apply_iagcwd('stretch_m2')
    imgs[name].apply_sigmoid_stretch()
    
    num_imgs += 1
    
    for key in imgs[name].img:
        if key in diff:
            diff[key] += imgs[name].show_diff(key, display = False, print_avg = False)
        else:
            diff[key] = imgs[name].show_diff(key, display = False, print_avg = False)

for key in diff:
    print(f'avg diff of {key} and Orig is: {diff[key] / num_imgs}')
```
